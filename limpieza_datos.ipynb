```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('premier_2017.csv')
```

```python
df.head()
```

```python
#quito varibles que de entrada no me sirven
df.drop(['Div','Referee'],axis=1,inplace=True)
```

```python
df.head()
```

```python
#resultado H:Local D:Empate A:Visitante
#renombrado columnas y poniendo resultado bien
df.columns=["Fecha","Local", "Visitante", "Goles_L","Goles_V","Resultado","Goles_Mitad_L","Goles_Mitad_V",
                   "Resultado_Mitad","Tiros_L","Tiros_V","Tiros_Buenos_L","Tiros_Buenos_V","Faltas_L","Faltas_V",
                  "Corners_L","Corners_V","Amarillas_L","Amarillas_V","Rojas_L","Rojas_V"]
df[['Resultado','Resultado_Mitad']]=df[['Resultado','Resultado_Mitad']].replace({"A": 'V', "D": 'E',"H":'L'})
```

```python
df.head()
```

```python
df.drop(['Goles_Mitad_L','Goles_Mitad_V','Resultado_Mitad'],axis=1,inplace=True)
df.head()
```

```python
#caso locales
variables=['Local','Fecha']+[i for i in df.columns if i.endswith('_L')]+[x for x in df.columns if x.endswith('_V')]+['Resultado']
df_final=pd.DataFrame()
df_final=df[variables]
df_final.columns=['Equipo','Fecha','Goles_0','Tiros_0','Tiros_Acertados_0','Faltas_0','Corners_0','Amarillas_0','Rojas_0',
                 'Goles_1','Tiros_1','Tiros_Acertados_1','Faltas_1','Corners_1','Amarillas_1','Rojas_1',
                 'Resultado']
df_final['Posicion']='L'
df_final.head()
```

```python
#caso Visitantes
variables=['Visitante','Fecha']+[i for i in df.columns if i.endswith('_V')]+[x for x in df.columns if x.endswith('_L')]+['Resultado']
df_final_2=pd.DataFrame()
df_final_2=df[variables]
df_final_2.columns=['Equipo','Fecha','Goles_0','Tiros_0','Tiros_Acertados_0','Faltas_0','Corners_0','Amarillas_0','Rojas_0',
                 'Goles_1','Tiros_1','Tiros_Acertados_1','Faltas_1','Corners_1','Amarillas_1','Rojas_1',
                 'Resultado']
df_final_2['Posicion']='V'
df_final_2.head()
```

```python
df_prueba=df_final.append(df_final_2)
```

```python
#y se gano o no??
y=[]
for i in range(0,len(df_prueba.Resultado)):
    if df_prueba['Resultado'].iloc[i]==df_prueba['Posicion'].iloc[i]:
        y.append(1)
    else:
        y.append(0)
df_prueba['y']=y
df_prueba['Fecha'] = pd.to_datetime(df_prueba['Fecha']) 
```

```python
df_prueba.sort_values(['Equipo', 'Fecha'],inplace=True)
```
